## Introduction

## Check VCF files

`bcftools view ./data/chr21.YRI.CEU.vcf.gz | less -S`

**without header**

`bcftools view ./data/chr21.YRI.CEU.vcf.gz -H | less -S`

## Extract data from a population

`bcftools view ./data/chr21.YRI.CEU.vcf.gz -S ./data/YRI.list | bgzip -c > ./data/chr21.YRI.vcf.gz`

`bcftools view ./data/chr21.YRI.CEU.vcf.gz -S ./data/CEU.list | bgzip -c > ./data/chr21.CEU.vcf.gz`

## Filter data

`bcftools view ./data/chr21.YRI.vcf.gz -f PASS -m 2 -M 2 -v snps | bgzip -c > ./data/chr21.YRI.biallelic.snps.vcf.gz`

`bcftools view ./data/chr21.CEU.vcf.gz -f PASS -m 2 -M 2 -v snps | bgzip -c > ./data/chr21.CEU.biallelic.snps.vcf.gz`
